# Agglomerative Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
```

```python
df = pd.read_csv('StoneFlakes.csv')
df
```

```python
# Drop the non-numeric column
df.drop(columns='ID',inplace=True)
```

```python
scaler = StandardScaler()
scaler.fit_transform(df)
```

```python
linkage_matrix = sch.linkage(df, method='complete')
```

```python
# plot dendogram
dendogram = sch.dendrogram(linkage_matrix)
plt.ylabel('height')
plt.savefig('dendogram.png', dpi=300)
```

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='complete')
```

```python
cluster_preds = cluster.fit_predict(df)
print(cluster_preds)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
X = df.iloc[:,[0,1]]
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
plt.scatter(x, y)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
y_kmeans = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=y_kmeans, cmap='brg')
plt.scatter(centroids[:,0], centroids[:,1], c='black', s=200, alpha=0.5)
```
