```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.mixture import GaussianMixture
```

```python
diab = load_diabetes(as_frame=True)
print(diab.DESCR)
```

```python
df = diab.data
```

```python
diab_df = df[['bmi','bp']]
```

```python
plt.xlabel('bmi')
plt.ylabel('bp')
plt.scatter(diab_df['bmi'],diab_df['bp'])
```

```python
em_gaussian = GaussianMixture(n_components=2)
cluster_preds = em_gaussian.fit_predict(diab_df)
```

```python
plt.xlabel('bmi')
plt.ylabel('bp')
plt.scatter(diab_df['bmi'],diab_df['bp'], c=cluster_preds, cmap='rainbow')
```

```python
print(em_gaussian.aic(diab_df))
print(em_gaussian.bic(diab_df))
```

```python
em_gaussian = GaussianMixture(n_components=5)
cluster_preds = em_gaussian.fit_predict(diab_df)
```

```python
plt.xlabel('bmi')
plt.ylabel('bp')
plt.scatter(diab_df['bmi'],diab_df['bp'], c=cluster_preds, cmap='rainbow')
```
