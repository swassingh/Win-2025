```python
## NN for regression
import urllib.request
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00243/yacht_hydrodynamics.data"
```

```python
raw_data = urllib.request.urlopen(url)
dataset = np.loadtxt(raw_data)
dataset
```

```python
print(dataset.shape)
```

```python
data_scale = StandardScaler().fit_transform(dataset)
```

```python
X_train,X_test,y_train,y_test = train_test_split(data_scale[:,0:6],data_scale[:,6],test_size=0.25,random_state=1)
```

```python
reg = MLPRegressor(hidden_layer_sizes=(3,6), random_state=1).fit(X_train,y_train)
```

```python
errors = abs(reg.score(X_test,y_test) / len(y_test))
errors
```

```python
## NN for classification
import sklearn
import urllib.request
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt"
```

```python
raw_data = urllib.request.urlopen(url)
dataset = pd.read_csv(raw_data, header=None, delim_whitespace=True)
dataset
```

```python
print(dataset.shape)
```

```python
data = dataset.dropna(axis=1, how='all')
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled, y, test_size=0.25,random_state=1)
```

```python
clf = MLPClassifier(
    solver='sgd',
    alpha=1e-5,
    hidden_layer_sizes=(5,3)
)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test, y_test))
```
