```python
# Define all states
all_states = [1,2,3,4,5,6]
```

```python
# Define rewards (treasure) for all states
rewards = {}
rewards[1] = -1
rewards[2] = 10
rewards[3] = 0
rewards[4] = 5
rewards[5] = 2
rewards[6] = -1
```

```python
# Dictionary of possible actions
actions = {
    2:(1,2,3),
    3:(2,3,4),
    4:(3,4,5),
    5:(4,5,6)
}
```

```python
# Define an initial policy
policy = {}
policy[2] = 1
policy[3] = 2
policy[4] = 5
policy[5] = 6
```

```python
# Define initial value function
V = {}
V[1] = -1
V[6] = -1
V[2] = 0
V[3] = 0
V[4] = 0
V[5] = 0
```

```python
# Hyperparameters
GAMMA = 0.9 # Discount factor (valuing future rewards less than immediate ones)
NOISE = 0.1 # Probability of taking a random action
SMALL_ENOUGH = 0.05  # Threshold for convergence
```

```python
import numpy as np

iteration = 0

while True:
  biggest_change = 0
  for s in all_states:
    if s in policy:
      old_v = V[s]
      new_v = 0

      for a in actions[s]:
        next = a

        # Choose a new random action to do
        act = np.random.choice([i for i in actions[s]])

        # Calculate the value
        v = rewards[s] + (GAMMA*((1-NOISE)*V[next] + (NOISE*V[act])))

        if v > new_v: # Is this better action? That's our new value then
          new_v = v
          policy[s] = a

      # Save the best of all actions for the state
      V[s] = new_v
      biggest_change = max(biggest_change, np.abs(old_v - V[s]))
  if biggest_change < SMALL_ENOUGH:
    break
  iteration += 1
  print(policy)
  print(iteration, V)
```
